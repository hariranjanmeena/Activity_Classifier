## Remove Unwanted line from  all batch files and adding label column with value equal to  respective folder name 

```python
import pandas as pd
```

```python
import os
```

```python
Classes = os.listdir('C:/Users/HP/Documents/ML projects/Project_Level_2/New folder (2)/Data/')
```

```python
for folder in Classes:
    # List of Files in old Folder
    files = os.listdir(f'Data/{folder}')
    print("pass",folder)

    for file_name in files:
        col_names = ['time','avg_rss12','var_rss12','avg_rss13','var_rss13','avg_rss23','var_rss23']
        df = pd.read_csv(f'Data/{folder}/{file_name}', skiprows=[0,1,2,3])
        df.columns= col_names
        df['label'] = folder
        df.to_csv(f'New data/{folder+file_name}')
        print("pass",file_name)
        
        
```

```python
for folder in Classes:
    
    # List of Files in old Folder
    all_filenames = os.listdir(f'New data/')
    extension = 'csv'
    #combine all files in the list
    combined_csv = pd.concat([pd.read_csv(f'New data/{f}') for f in all_filenames ])
    #export to csv
    combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
        
        
```
